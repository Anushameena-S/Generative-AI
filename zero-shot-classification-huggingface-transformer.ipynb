The pipeline function in the Transformers library (by Hugging Face) is a high-level API that allows you to easily perform various natural language processing (NLP) and vision-related tasks without needing to dive into the complexity of model initialization, tokenization, and inference.
Pipelined Model , this function will automatically load a pretrained model suited for the Task.

```python
from transformers import pipeline
```

```python
text_classifier = pipeline("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
sequence_to_classify = "It has been raining heavily througout the day but there is no flood."

candidate_labels = ["weather","rain","cyclone","music","travel","negative"]
```

```python
text_classifier(sequence_to_classify, candidate_labels)
```

```python
sequence_to_classify = "It has been raining heavily througout the day but there is no flood."

candidate_labels = ["weather-news","entertainment","travel","negative"]
```

```python
text_classifier(sequence_to_classify, candidate_labels)
```

```python
import pandas as pd
```

```python
headlines = pd.read_csv("/kaggle/input/india-news-headlines/india-news-headlines.csv")
```

```python
headlines.head()
```

```python
headlines.shape
```

```python
#random_news_sample = headlines.sample(n = 100, random_state = 42)
random_news_sample = headlines.sample(n = 100)
```

```python
sample = list(random_news_sample.headline_text.values)
```

```python
news_candidate_labels = ["politics","leadership","business and economy","crime and law","sports","entertainment","technology", 
                    "health", "environment and weather"]
```

```python
output = text_classifier(sample, news_candidate_labels)
```

```python
output_df = pd.DataFrame(output)
```

```python
output_df.head()
```

```python
output_df['predicted_label'] = output_df['labels'].apply(lambda val : val [0])
```

```python
output_df['predicted_scores'] = output_df['scores'].apply(lambda val : val [0])
```
