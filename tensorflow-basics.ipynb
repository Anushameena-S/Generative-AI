#### 5. Mathematical operation in tensor

```python
X = tf.constant([[2],
                 [3],
                 [4],
                 [5],
                 [6],
                 [3]], dtype = 'float16')
```

```python
Weights = tf.Variable([[0.1],
                       [0.3],
                       [0.17],
                       [0.5],
                       [0.6],
                       [1.3]], dtype = 'float16')
```

```python
WT = tf.transpose(Weights)
```

```python
# Y = WT * X + Bias
Y = tf.matmul(WT, X)
Y
```

```python
bias=tf.constant(2.4, dtype='float16')
```

```python
tf.matmul(WT, X) + bias
```

#### Dense Neural Network / Feed Forward Network / Fully connected Layer - Multi Classification

```python
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Sequential 
from tensorflow.keras.optimizers import Adam, Adagrad
```

```python
def build_model():
    model = Sequential()
    # input shape = num of features
    model.add(Input(shape = (8,)))
    # Input Layer - I
    # units = number of neurons in each layer, activation function = 'relu'
    model.add(Dense(units = 32, activation = 'relu'))
    # Hidden Layer - I
    model.add(Dense(units = 64, activation = 'relu'))
    # Hidden Layer - II
    model.add(Dense(units = 128, activation = 'relu'))
    # Ouput Layer
    model.add(Dense(units = 3, activation = 'softmax'))
    model.compile(loss = 'categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
    return model
```

```python
model = build_model()
```
